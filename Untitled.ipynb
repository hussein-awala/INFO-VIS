```python
additives=pd.read_csv('tsv/products_additives.tsv','\t')
```

```python
country_products=pd.read_csv('tsv/products_countries.tsv','\t')
```

```python
additives.head()
```

```python
country_products.head()
```

```python
d=country_products.merge(additives,left_on='code',right_on='code',how='inner')
```

```python
d.head()
```

```python
d1=country_products.groupby('country').count()
d1.columns=['nbOfProduct']
```

```python
d2=d.loc[:,['country','code']].groupby('country')[['code']].nunique()
d2.columns=['nbOfAdditive']
```

```python
d3=d1.merge(d2,left_index=True,right_index=True,how='left').fillna(0)
```

```python
d3.head()
```

```python
d1.head()
```

```python
d3.to_csv('tsv/dataset.tsv','\t')
```
